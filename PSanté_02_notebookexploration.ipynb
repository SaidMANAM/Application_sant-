### La distribution des variables
Pour nos variables numériques on va afficher leurs histogrammes pour avoir une idée de la distribution de chacune des varibales


```python
def histogram(data, name):
    BIGGER_SIZE=15
    plt.rc('font', size=BIGGER_SIZE)# taille de texte par défaut
    plt.rc('axes', titlesize=BIGGER_SIZE)# taille des titres des axes
    plt.rc('axes', labelsize=BIGGER_SIZE)# taille des labels des axes
    plt.rc('xtick', labelsize=BIGGER_SIZE)# taille des ticks des ascisses
    plt.rc('ytick', labelsize=BIGGER_SIZE)# taille des ticks des ordonnées
    plt.rc('figure', titlesize=BIGGER_SIZE)# taille du titre
    #plt.subplot(2,3,i)
    fig, ax = plt.subplots(1, figsize=(12,8))
    #fig, ax =plt.figure(figsize=(10,6))
    quant_25, quant_50, quant_75 =  data.quantile(0.25), data.quantile(0.5), data.quantile(0.75)
    n, bins, patches=plt.hist(data,color='c',edgecolor='black', linewidth=1.2 )
    quants = [[quant_25, 0.8, 0.26], [quant_50, 1, 0.36],  [quant_75, 0.8, 0.46]]
    ax.text(quant_25-.13, 0.27, "25th", size = 11, alpha = 0.85)
    ax.text(quant_50-.13, 0.37, "50th", size = 12, alpha = 1)
    ax.text(quant_75-.13, 0.47, "75th", size = 11, alpha = 0.85)
    ax.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    for i in quants:
        ax.axvline(i[0], alpha = i[1], ymax = i[2],color='k', linestyle = 'dashed', linewidth=1)
    #xint = []
    #locs, labels = plt.xticks()
   # for each in locs:
        #xint.append(int(each))
    #plt.xticks(xint)
    #plt.xlim([0,data.shape[1]])
    #plt.yticks([0,20,40,60,80,100,120,125])
    # plot values on top of bars
    xticks = [(bins[idx+1] + value)/2 for idx, value in enumerate(bins[:-1])]
    for idx, value in enumerate(n):
        if value > 0:
           plt.text(xticks[idx], value+2, int(value), ha='center')

    plt.ylabel("Nombre de produits")
    plt.xlabel("Valeur de la variable")
    plt.grid()
    
    plt.title("La distribution de la variable: "+elt)
    plt.show()
   
num_data=df._get_numeric_data().drop('nova_group',axis=1)
print(num_data.columns)
for elt in num_data :
    histogram(df[elt],elt) 
```

#### Tableau de contingence

Le tableau de contingence est une méthode de représentation de données issues d’un comptage permettant d'estimer la dépendance entre deux caractères. Elle consiste à croiser deux caractères d'une population (par exemple une classe 'nutriscore' et 'nova_group')en dénombrant l'effectif correspondant à la conjonction « caractère 1 » et « caractère 2» 

```python
data_crosstab = pd.crosstab(df['nutriscore_grade'], 
                            df['nova_group'],  
                               margins = True,margins_name='Total') 
#aggfunc={'Total':np.mean}
plt.figure(figsize=(10,6))
print(data_crosstab) 
data_crosstab_Sans_Total = pd.crosstab(df['nutriscore_grade'], 
                            df['nova_group'],  
                               ) 
sns.heatmap(data_crosstab_Sans_Total,linewidths=.5, cmap="YlGnBu",annot=True, linecolor='black' )

plt.title('Carte de chaleur du nutriscore_grade et nova_group ')
plt.show()
```

## Analyse multivariée

### Réduction de dimensionnalité via  Analyse des Composantes principales

Dans cette partie on va ffectuer une analyse descriptive avec ACP et une analyse explicative avec l'ANOVA

```python
#On va centrer et réduire les données avant ACP
features=df_scaled.columns
df_scaled=sklearn.preprocessing.StandardScaler().fit_transform(df_scaled)
len(features)
# On vérifie que la normalisation est un succès
print("Moyennes :\n", np.mean(df_scaled, axis = 0)) # Les moyennes devraient valoir 0 ou être infiniment petites avec l'arrondi
print("\nÉcarts-types :\n", np.std(df_scaled, axis = 0)) # Les écarts-types devraient valoir 1
print("\nMatrice de corrélation :\n", (1/df_scaled.shape[0]) * np.matmul(np.transpose(df_scaled), df_scaled)) # La matrice doit avoir des 1 sur la diagonale
```

```python
#from math import round
acp = PCA()
acp.fit_transform(df_scaled)
#pd.DataFrame(df_scaled,index=df._get_numeric_data().index, columns=['F'+str(i+1) for i in range(6)])
# Éléments nécessaire au choix du nombre de composantes principales
print("Valeurs propres :\n", acp.explained_variance_)#La quantité de variance expliquée par chacun des composants sélectionnés. Égal aux plus grandes valeurs propres de n_components de la matrice de covariance de X
print("\nQualité de représentation des axes en % :\n", acp.explained_variance_ratio_ * 100)
print("\nQualité cumulée en % :\n ", np.round(np.cumsum(acp.explained_variance_ratio_)* 100,1))
```

On observera une contribution de 84.2% lorsqu'on ne considère que les cinq premiers axes, ceci grace au critère de'Elbow';  On ne retiendra que les axes associés à des valeurs propre supérieures à 1

### Analyse de la variance 'ANOVA'

l'analyse de la variance ( ANOVA ) est une technique utilisée pour vérifier si les moyennes des groupes proviennent d'une même population. Les groupes correspondent aux modalités d'une variable qualitative (p. ex. variable : nova_group; groups_Pnns) et les moyennes sont calculés à partir d'une variable continue (p. ex. nutriscore_score).

#### L'analyse des variables: 'nova_group' et 'nutriscore_score'

```python
Box_Plot_cat(groupes, 'nova_group', modalites,'nutriscore_score',vert=True)
```

Les 4 variétés semblent assez différentes, même si l'ordre de grandeur de ces écarts n'est pas très grand. La question sera de savoir si ces écarts sont significatifs ou pas.

C'est l'ANOVA qui nous permettra de répondre à cette question.

#####  Test de Fisher

```python
anova_variete = smf.ols('nutriscore_score~nova_group', data=df).fit()
print(anova_variete.summary())
sm.stats.anova_lm(anova_variete, typ=2)
```



- l'hypothèse H0 : Les 'nova_group' n'ont pas d'effet sur le nutriscore 
                           
- Ce qui nous intéresse réellement, c'est le test de Fisher. La p-valeur de ce test est nulle => On rejette donc l'hypothèse H0.

- le 'nova_group' a donc bien un effet sur le nutriscore
       

#### L'analyse des variables: 'pnns_groups' et 'nutriscore_score'

```python
fig = plt.figure(figsize=(12,8))
ax = sns.boxplot(data=df, x='pnns_groups_1', y='nutriscore_score')
plt.setp(ax.get_xticklabels(), rotation=90)
plt.title("Influence de la catégorie de produit sur le Nutriscore", fontdict=font_title)
plt.grid()
plt.show()
```

On remarque ici ques les catégories semblent assez différentes, même si l'ordre de grandeur des écart est relativement faible. Dans les histogrammes ci-dessus, nous voyons cependant que les distributions ne semblent pas suivre la loi normale.
La question sera à présent de savoir si ces écarts sont significatifs ou pas via l'analyse de variance :

#####  Test de Fisher

```python
anova_category = smf.ols('nutriscore_score~pnns_groups_1', data=df).fit()
print(anova_category.summary())
sm.stats.anova_lm(anova_category, typ=2)
```

Les résultats du test de Fisher nous indiquent ici une p-value de 0 pour l'ensemble des catégories, donc inferieur au niveau de test de 5%. Nous rejettons donc l'hypothèse H0 selon laquelle les ditributions sont identiques.
La catégorie de produit a donc bien une influence sur le Nutriscore.
